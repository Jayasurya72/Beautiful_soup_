```python
#importing the required libraries
from bs4 import BeautifulSoup
import requests
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
#first,from html tag we should extract an first title.
first_title=soup.find('h2',class_="title")
first_title
```

```python
#print first job title
first_title.text
```

```python
#Second,from html tag we should extract the first company name

first_company=soup.find('span',class_="company")
first_company
```

```python
first_company.text
```

```python
first_company.text.replace('\n',"")
```

```python
#third,from html tag we should extract an how many days earlier an first job is posted.

day=soup.find('span',class_="date date-a11y")
day
```

```python
day.text
```

```python
#forth, from html tag we should extract an location of first job

location=soup.find('span',class_="location accessible-contrast-color-location")
location
```

```python
location.text
```

# Now  we will scrap for multiple job position

```python
# first,from html tag we will extract job title of all.

titles=soup.find_all('h2',class_="title")
titles
```

```python
#for all job tiles we cannot able execute text by using text comment.
#so we will extract the text from the tags one by one by using looping over tags.

job_titles=[]
for i in titles:
    job_titles.append(i.text)
job_titles
```

```python
job_titles[5]
```

```python
job_titles=[]
for i in titles:
    job_titles.append(i.text.replace('\n',""))
job_titles
```

```python
# Second,from html tag we will extract company names of all.

company_names=soup.find_all('span',class_="company")
company_names
```

```python
#for all company names we cannot able execute text by using text comment.
#so we will extract the text from the tags one by one by using looping over tags.

job_company_names=[]
for i in company_names:
    job_company_names.append(i.text)
job_company_names
```

```python
job_company_names=[]
for i in company_names:
    job_company_names.append(i.text.replace('\n',""))
job_company_names
```

```python
job_company_names[5]
```

```python
# third,from html tag we will extract how many days earlier the job is posted of all the companies .

days=soup.find_all('span',class_="date date-a11y")
days
```

```python
#for all the days we cannot able execute text by using text comment.
#so we will extract the text from the tags one by one by using looping over tags.

job_days=[]
for i in days:
    job_days.append(i.text)
job_days
```

```python
location_names=soup.find_all('span',class_="location accessible-contrast-color-location")
location_names
```

```python
job_location_names=[]
for i in location_names :
    job_location_names.append(i.text)
job_location_names
```

```python
#use if and else
if i.text==None:
    job_location_names.append("-")
else:
    job_location_names.append(i.text)
job_location_names
```

```python
len(job_location_names)
```

```python
len(job_titles)
```

```python
print(len(job_titles),len(job_company_names),len(job_days),len(job_location_names))
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['Title']=job_titles
jobs['Companies']=job_company_names
jobs['Days before posted']=job_days
jobs['Location']=job_location_names
```
